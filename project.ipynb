```python
# Select columns with numerical data types
numerical_columns = df.select_dtypes(include=['number']).columns

# Display the numerical columns
print(numerical_columns)
print(df.describe())
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns

# Display the categorical columns
print(categorical_columns)

for col in categorical_columns:
    print('_'*20)
    print(f"{col}: {df[col].nunique()}")
    print(f"{col}: {df[col].unique()}")
```

```python
# Convert 2 dates columns to datetime type:
daytime_columns = ['Date of Admission','Discharge Date']
for x in daytime_columns:
    df[x] = pd.to_datetime(df[x])
```

```python
# Convert the Name to Capitalize each word:
df.Name = df.Name.str.title()
```

```python
#Check the results
print(df.info())
print(df.Name.head(10))
```

```python
#Exploratory Data Analysis (EDA)

#List out columns with unique values less than 10:

less_than_10_categories = []

for col in categorical_columns:
    if df[col].nunique() < 10:
        less_than_10_categories.append(col)

print(less_than_10_categories)
```

```python
for col in less_than_10_categories:
    print('='*50)
    print(df[col].value_counts().sort_index())
    plt.figure(figsize=(10,3))
    sns.barplot(x = df[col].sort_values().unique(),y = df.groupby(col).Name.count())
    plt.title(f'Value counts per {col}')
    plt.ylabel('Value counts')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le_dict = {}
mapping_dict = {}
# Encode each categorical column using LabelEncoder
for col in less_than_10_categories:
    le = LabelEncoder()
    df[col + '_encoded'] = le.fit_transform(df[col])
    le_dict[col] = le
    mapping_dict[col] = dict(enumerate(le.classes_))

print("Label Encoding Mappings:")
for col, mapping in mapping_dict.items():
    print(f"{col}: {mapping}\n")
    
print("Encoded DataFrame:")
df.head()
```

```python
# Number of rows needed (rounded up)
n_rows = int(np.ceil(len(numerical_columns) / 3))

# Create subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=3, figsize=(15, 5 * n_rows))

# Flatten the axes array for easy iteration (in case it's a 2D array)
axes = axes.flatten()

# Iterate over numerical columns and axes to create histograms
for i, col in enumerate(numerical_columns):
    sns.histplot(df[col], ax=axes[i])
    axes[i].set_title(f'Histogram of {col}')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  # Adjusts subplots to fit into the figure area.
plt.show()
```
